```python
#Skewness and Pearson’s Median Skewness Coefficient
```

```python
def skewness(x):
    res = 0
    m = x.mean()
    s = x.std()
    for i in x :
        res += (i-m)*(i-m)*(i-m) 
    res /= (len(x)*s*s*s)
    return res

print(f"Skewness of the male population : {skewness(ml2_age)}")
print(f"Skewness of the female population : {skewness(fm2_age)}")
```
