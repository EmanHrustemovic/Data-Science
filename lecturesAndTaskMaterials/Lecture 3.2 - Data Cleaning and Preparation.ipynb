```python
datas = pd.DataFrame({'food' : ['bacon','pulled_pork','bacon',
                               'Pastraml','corned beef','Bacon',
                               'pastraml','honey ham','nova lox'],
                     'ounces' : [4,3,12,6,7.5,8,3,5,6]})

meat_to_animal = {
    'bacon' : 'pig',
    'pulled pork' : 'pig',
    'pastraml' : 'cow',
    'corned beef' : 'cow',
    'honey ham' : 'pig',
    'nova lox' : 'salmon'
}
```

```python
lowercased = datas['food'].str.lower()
lowercased
```

```python
datas['animal'] = lowercased.map(meat_to_animal)
```

```python
datas
```
