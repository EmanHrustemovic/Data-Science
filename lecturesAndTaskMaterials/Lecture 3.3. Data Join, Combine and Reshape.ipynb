```python
#pandas.merge
```

```python
df3 = pd.DataFrame({'key' : ['b','b','a','c','a','a','b'],
                   'data1' : range(7)})

df4 = pd.DataFrame({'key' : ['a','b','d'],
                   'data1' : range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,on='key')
```

```python
df5 = pd.DataFrame({'lkey' : ['b','b','a','c','a','a','b'],
                   'data1' : range(7)})

df6 = pd.DataFrame({'rkey' : ['a','b','d'],
                   'data1' : range(3)})
```

```python
df5
```

```python
df6
```

```python
pd.merge(df5,df6,left_on='lkey',right_on='rkey')
```

```python
pd.merge(df5,df6,how='outer')
```

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','b'],
                   'data' : range(6)})

df2 = pd.DataFrame({'key':['a','b','a','b','d'],
                   'data' : range(5)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='key',how='left') #many-to-many merge
```

```python
pd.merge(df1,df2,how='inner')
```
